```python
# This command imports the pandas library
#which is used for data manipulation and analysis, and assigns it the alias
import pandas as pd
```

```python
#This command reads a CSV file from your Google Drive and
#loads it into a pandas DataFrame named df
df = pd.read_csv('/content/drive/MyDrive/Byrne Dataset/synthetic_nursing_notes.csv')
```

```python
#This command displays the first 5 rows of the df DataFrame....
#allowing you to get a preview of the data
df.head()
```

```python
# Calculate the mean of the 'age' column
age_mean = df['age'].mean()
print(f"Mean of age: {age_mean}")

# Calculate the counts for categorical columns
gender_counts = df['gender'].value_counts()
race_counts = df['race'].value_counts()
ethnicity_counts = df['ethnicity'].value_counts()
marital_status_counts = df['marital_status'].value_counts()

print("\nCounts for gender:")
display(gender_counts)

print("\nCounts for race:")
display(race_counts)

print("\nCounts for ethnicity:")
display(ethnicity_counts)

print("\nCounts for marital_status:")
display(marital_status_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a box plot for the 'age' column
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['age'])
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a cross-tabulation of gender and race
gender_race_counts = pd.crosstab(df['gender'], df['race'])

# Create a heatmap of the counts
plt.figure(figsize=(10, 7))
sns.heatmap(gender_race_counts, annot=True, fmt='d', cmap='Blues')
plt.title('Counts of Gender vs. Race')
plt.xlabel('Race')
plt.ylabel('Gender')
plt.show()

# Create a cross-tabulation of ethnicity and marital status
ethnicity_marital_counts = pd.crosstab(df['ethnicity'], df['marital_status'])

# Create a heatmap of the counts
plt.figure(figsize=(10, 7))
sns.heatmap(ethnicity_marital_counts, annot=True, fmt='d', cmap='Blues')
plt.title('Counts of Ethnicity vs. Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Ethnicity')
plt.show()
```
